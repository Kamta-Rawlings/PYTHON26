Regex is a way to search and validate or extract partens in text
Example: To find emails , extract numbers , validate passwords and clean messy data
To begin using regex, start with import re

```python
import re
```

re.search() is used to find first match 
Example:

```python
import re
text ="My 56789 number is 1234"
match = re.search(r"\d+",text)
print(match.group())
# We use match.group() to join all the numbers after finding the match
# 
```

re.findall => An example to find all the matches

```python
findall = re.findall(r"\d+", text)
print(findall)
```

An example to replace (sub) => re.sub

```python
sub = re.sub(r"\d+","X", text)
print(sub)
```

Some Regex Characters you should Master : CHARACTERS
| Pattern | Meaning                    |
| ------- | -------------------------- |
| `.`     | any character              |
| `\d`    | digit (0–9)                |
| `\w`    | letter, number, underscore |
| `\s`    | whitespace                 |
| `\D`    | Each non digits character  | 
| `\b`    |Word Boundary(Start Of Word)|

Quantifiers
| Pattern | Meaning         |
| ------- | --------------- |
| `+`     | one or more     |
| `*`     | zero or more    |
| `?`     | optional        |
| `{n}`   | exactly n times |
| `{n,m}` | between n and m |
| `( )`   | Capturing group |


Anchors
| Pattern | Meaning         |
| ------- | --------------- |
| `^`     | start of string |
| `$`     | end of string   |


Sets
`[abc]`     # a or b or c
`[a-z]`     # lowercase letters
`[0-9]`   # digits
`[^abc]`  # NOT a, b, c

Example: To extract Emails

```python
import re 
text = "Contact me at test@gmail.com"
re.findall(r"\w+@\w+\.\w+",text)
```

Example: To Extract Numbers

```python
text2 = "My phone number is +32680184013"
re.findall(r".\d+",text2)
```

Valid Password(Basic)

```python
password = "abc123"

if re.search(r"\d", password):
    print("Contains number")
```
